```python
def split_data(df):

    # Create empty DataFrames for each split
    df_train = pd.DataFrame(columns=df.columns)
    df_test = pd.DataFrame(columns=df.columns)
    df_val = pd.DataFrame(columns=df.columns)

    # Shuffle original DataFrame to ensure randomness
    df = df.sample(frac=1, random_state=42).reset_index(drop=True)

    # Group and split
    for fault_class, group in df.groupby('anomaly_class'):
        total = len(group)
        test_size = int(total * 0.14)
        val_size = int(total * 0.01)
        train_size = total - test_size - val_size

        # Shuffle group rows
        group = group.sample(frac=1, random_state=42).reset_index(drop=True)

        # Split the group
        test_split = group.iloc[:test_size]
        val_split = group.iloc[test_size:test_size + val_size]
        train_split = group.iloc[test_size + val_size:]

        # Append to final DataFrames
        df_train = pd.concat([df_train, train_split], ignore_index=True)
        df_test = pd.concat([df_test, test_split], ignore_index=True)
        df_val = pd.concat([df_val, val_split], ignore_index=True)

    print("Train:", len(df_train), "Test:", len(df_test), "Val:", len(df_val))

    return df_train, df_test, df_val
```

```python
class InfraredDataset(Dataset):
    def __init__(self, dataframe, classification_type=classification_type,transform=None, label_map=None):
        self.df = dataframe.reset_index(drop=True)
        self.transform = transform
        self.classification_type = classification_type
        self.label_map = label_map or {label: idx for idx, label in enumerate(self.df[classification_type].unique())}

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        img_path = self.df.loc[idx, 'image_filepath']
        label = self.label_map[self.df.loc[idx, self.classification_type]]

        # Load image in grayscale (as NumPy array)
        image = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)

        if image is None:
            raise ValueError(f"Image not found or failed to load: {img_path}")

        # Convert the NumPy array to a PIL Image
        image = Image.fromarray(image)

        # Apply transform if available
        if self.transform:
            image = self.transform(image)  # Apply transformation (e.g., Resize, Normalize)
        else:
            # Default: convert to torch tensor
            image = torch.from_numpy(image).float() / 255.0

        return image, label
```

```python
df_train, df_test, df_val = split_data(df)

train_transform = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),  # optional, tweak for IR variation
    transforms.ToTensor(),                                # If not already a tensor
    transforms.Normalize(mean=[0.5], std=[0.5])
])

val_test_transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5], std=[0.5])
])

# Pass your DataFrame and transform
train_dataset = InfraredDataset(df_train,transform=train_transform)
val_dataset = InfraredDataset(df_val,transform=val_test_transform)
test_dataset = InfraredDataset(df_test,transform=val_test_transform)

# Prepare dataset on PyTorch DataLoader
train_loader = DataLoader(
    train_dataset,
    batch_size=32,
    shuffle=True,        # Shuffle only for training
    pin_memory=True      # Boost speed if using GPU
)

val_loader = DataLoader(
    val_dataset,
    batch_size=32,
    shuffle=False,
    pin_memory=True
)

test_loader = DataLoader(
    test_dataset,
    batch_size=32,
    shuffle=False,
    pin_memory=True
)
```

## Define the ssc gan model

### Subtask:
Define the architecture for the generator and discriminator networks of the SSC GAN, incorporating the supervised branch for classification.


**Reasoning**:
Define the Generator and Discriminator classes with the specified architecture and instantiate them.



```python
class Generator(nn.Module):
    def __init__(self, latent_dim, img_shape):
        super(Generator, self).__init__()
        self.img_shape = img_shape

        self.model = nn.Sequential(
            nn.Linear(latent_dim, 128 * 8 * 8),
            nn.BatchNorm1d(128 * 8 * 8),
            nn.LeakyReLU(0.2, inplace=True),
            nn.Unflatten(1, (128, 8, 8)),

            nn.ConvTranspose2d(128, 128, 4, stride=2, padding=1),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2, inplace=True),

            nn.ConvTranspose2d(128, 64, 4, stride=2, padding=1),
            nn.BatchNorm2d(64),
            nn.LeakyReLU(0.2, inplace=True),

            nn.ConvTranspose2d(64, img_shape[0], 4, stride=2, padding=1),
            nn.Tanh()
        )

    def forward(self, z):
        img = self.model(z)
        return img.view(img.size(0), *self.img_shape)

class Discriminator(nn.Module):
    def __init__(self, img_shape, num_classes):
        super(Discriminator, self).__init__()

        self.conv_layers = nn.Sequential(
            nn.Conv2d(img_shape[0], 32, 4, stride=2, padding=1),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(32, 64, 4, stride=2, padding=1),
            nn.BatchNorm2d(64),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(64, 128, 4, stride=2, padding=1),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2, inplace=True),
        )

        # Adversarial branch
        self.adversarial_layer = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128 * 8 * 8, 1),
            nn.Sigmoid()
        )

        # Supervised branch
        self.supervised_layer = nn.Sequential(
            nn.Flatten(),
            nn.Linear(128 * 8 * 8, num_classes),
            nn.Softmax(dim=1)
        )

    def forward(self, img):
        features = self.conv_layers(img)
        adversarial_output = self.adversarial_layer(features)
        supervised_output = self.supervised_layer(features)
        return adversarial_output, supervised_output

# Instantiate models
latent_dim = 100
img_height, img_width = 64, 64
img_shape = (1, img_height, img_width) # Grayscale image

generator = Generator(latent_dim, img_shape)
discriminator = Discriminator(img_shape, num_classes)

# Print model architectures
print("Generator Architecture:")
print(generator)
print("\nDiscriminator Architecture:")
print(discriminator)
```

## Set up training parameters

Define the loss functions (including GAN loss and classification loss), optimizers, and other training parameters for the SSC GAN.

## Train the ssc gan

### Subtask:
Implement the training loop to train the generator and discriminator networks using primarily the normal data, while also utilizing the classification labels for the supervised branch.

```python
# Set models to training mode
generator.train()
discriminator.train()

# Adversarial ground truths and false
valid = torch.ones((train_loader_normal.batch_size, 1), device=device)
fake = torch.zeros((train_loader_normal.batch_size, 1), device=device)

# Label for the normal class (used for supervised loss on generated images)
normal_class_label = label_map_12_classes['No-Anomaly']
real_normal_labels = torch.full((train_loader_normal.batch_size,), normal_class_label, dtype=torch.long, device=device)


# Lambda for balancing supervised loss
lambda_supervised = 1.0 # Adjust as needed

# Training loop
for epoch in range(num_epochs):
    for i, (imgs, _) in enumerate(train_loader_normal): # We only need images from normal data

        # Transfer images to device
        real_imgs = imgs.to(device)

        # ---------------------
        #  Train Discriminator
        # ---------------------

        optimizer_D.zero_grad()

        # Adversarial loss for real images
        real_validity, real_supervised = discriminator(real_imgs)
        d_real_adv_loss = adversarial_loss(real_validity, valid[:real_imgs.size(0)])

        # Supervised loss for real images (only for normal class)
        # Use only the first dimension of real_supervised which corresponds to the batch size
        d_real_supervised_loss = supervised_loss(real_supervised[:real_imgs.size(0), :], real_normal_labels[:real_imgs.size(0)])


        # Generate fake images
        z = torch.randn((real_imgs.size(0), latent_dim), device=device)
        fake_imgs = generator(z)

        # Adversarial loss for fake images
        fake_validity, fake_supervised = discriminator(fake_imgs.detach()) # detach to not update generator
        d_fake_adv_loss = adversarial_loss(fake_validity, fake[:real_imgs.size(0)])

        # Total discriminator loss
        d_loss = d_real_adv_loss + d_fake_adv_loss + lambda_supervised * d_real_supervised_loss

        # Backprop and update
        d_loss.backward()
        optimizer_D.step()

        # -----------------
        #  Train Generator
        # -----------------

        optimizer_G.zero_grad()

        # Generate fake images (need to generate again for generator training)
        z = torch.randn((real_imgs.size(0), latent_dim), device=device)
        fake_imgs = generator(z)

        # Adversarial and supervised loss for fake images (generator wants discriminator to classify them as real and normal)
        fake_validity, fake_supervised = discriminator(fake_imgs)
        g_adv_loss = adversarial_loss(fake_validity, valid[:real_imgs.size(0)])
        g_supervised_loss = supervised_loss(fake_supervised, real_normal_labels[:real_imgs.size(0)])

        # Total generator loss
        g_loss = g_adv_loss + lambda_supervised * g_supervised_loss

        # Backprop and update
        g_loss.backward()
        optimizer_G.step()

        # ---------------
        # Log Progress
        # ---------------
        if (i+1) % 100 == 0 or (i+1) == len(train_loader_normal):
            print(f"Epoch [{epoch+1}/{num_epochs}] Batch [{i+1}/{len(train_loader_normal)}] "
                  f"D Loss: {d_loss.item():.4f} (Real Adv: {d_real_adv_loss.item():.4f}, Fake Adv: {d_fake_adv_loss.item():.4f}, Real Sup: {d_real_supervised_loss.item():.4f}) "
                  f"G Loss: {g_loss.item():.4f} (Adv: {g_adv_loss.item():.4f}, Sup: {g_supervised_loss.item():.4f})")
```
